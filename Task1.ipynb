```python
import pandas as pd
import numpy as np
import ast
```

```python
df = pd.read_csv("Retail_Transactions_Dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df['Promotion'] = df['Promotion'].fillna('No Promotion')
```

```python
df['Product_List'] = df['Product'].apply(ast.literal_eval)
df['Num_Products'] = df['Product_List'].apply(len)
```

```python
df['City'] = df['City'].str.title()
df['Store_Type'] = df['Store_Type'].str.title()
```

```python
df['Avg_Price_Per_Item'] = df['Total_Cost'] / df['Total_Items']
```

```python
df.to_csv("Retail_Transactions_Cleaned.csv", index=False)
```
